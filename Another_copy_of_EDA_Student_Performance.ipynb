```python

import matplotlib.pyplot as plt
import seaborn as sns

sns.barplot(x="Gender", y="GPA", data=df)
plt.title("Gender vs GPA")
plt.show()
```

```python
sns.barplot(x="ParentalEducation", y="GPA", data=df)
plt.xticks(rotation=45)
plt.title("Parental Education vs GPA")
plt.show()
```

```python
sns.scatterplot(x="StudyTimeWeekly", y="GPA", data=df)
plt.title("Study Time Weekly vs GPA")
plt.show()
```
